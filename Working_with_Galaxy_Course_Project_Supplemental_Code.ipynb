Downloaded/copied data from SnpEff Eff HTML sheets ("Gene information") into excel CSV. Then uploaded here as pandas dataframes to group, count, and sort. 

```python
import pandas as pd
```

```python
# Daughter
df_80 = pd.read_csv('NA12880_annotated.csv')
df_80.head()
```

```python
df_80['name'] = df_80['#GeneName'].apply(lambda x: x.split()[0])
df_80.head()
```

```python
grouped = df_80.groupby(by='name').count()
df_80_sorted = grouped.sort_values(by='#GeneName', ascending=False)
df_80_sorted.head()
```

```python
# Mother
df_78 = pd.read_csv('NA12878_annotated.csv')
df_78.head()
```

```python
grouped = df_78.groupby(by='#GeneName').count()
df_78_sorted = grouped.sort_values(by='GeneId', ascending=False)
df_78_sorted.head()
```

```python
# Father 
df_77 = pd.read_csv('NA12877_annotated.csv')
df_77.head()
```

```python
grouped = df_77.groupby(by='#GeneName').count()
df_77_sorted = grouped.sort_values(by='GeneId', ascending=False)
df_77_sorted.head()
```
